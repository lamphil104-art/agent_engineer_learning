# Giáo trình AI Engineer để build AI Agent theo từng ngày

Đây là giáo trình [curriculum] mình sẽ dạy bạn theo kiểu từng ngày, đi từ nền tảng đến lúc tự build được AI agent [tác nhân AI] khá hoàn chỉnh.

Mình thiết kế theo hướng:
- thực chiến [hands-on]
- bám background [nền tảng] BI / Data Engineer của bạn
- ưu tiên Gemini + Python + VS Code + PowerShell
- học để build được project thật, không chỉ học lý thuyết

---

# Tổng quan lộ trình

Giáo trình này chia thành 5 chặng [phases]:

## Chặng 1: Foundation [nền tảng]
Mục tiêu:
- hiểu LLM app [ứng dụng LLM]
- biết gọi API
- biết prompt tốt
- biết structured output [đầu ra có cấu trúc]

## Chặng 2: RAG Basics [nền tảng RAG]
Mục tiêu:
- đọc tài liệu
- chia chunk [đoạn]
- embedding [biểu diễn vector]
- retrieval [truy xuất]

## Chặng 3: Tool-Using AI [AI dùng công cụ]
Mục tiêu:
- agent biết gọi tool [công cụ]
- route [định tuyến] task
- workflow [luồng] nhiều bước

## Chặng 4: Frameworks [khung phát triển]
Mục tiêu:
- LangChain
- LangGraph
- hiểu framework làm gì, không phụ thuộc mù quáng

## Chặng 5: Production [triển khai thực tế]
Mục tiêu:
- evaluation [đánh giá]
- logging / tracing [theo dõi]
- guardrails [rào chắn an toàn]
- deploy [triển khai]
- portfolio projects

# Giáo trình theo từng ngày

## Day 1
### Chủ đề:
Setup môi trường [environment setup] + gọi Gemini API đầu tiên

### Học gì:
- AI agent khác chatbot thế nào
- tạo `.venv`
- cài package
- dùng `.env`
- gọi Gemini API
- tạo mini app tóm tắt text thành JSON

### Kết quả cuối buổi:
- chạy được `main.py`
- có `outputs/result.json`

---

## Day 2
### Chủ đề:
Prompt Engineering [kỹ thuật viết prompt] + Structured Output nâng cao

### Học gì:
- prompt tốt vs prompt kém
- role / task / rules / output format
- mở rộng schema JSON
- trích xuất `topic`, `summary`, `key_points`, `sentiment`, `action_items`

### Kết quả cuối buổi:
- mini app phân tích text có cấu trúc tốt hơn
- hiểu cách ép model trả output ổn định

---

## Day 3
### Chủ đề:
Code organization [tổ chức mã nguồn] + reusable functions [hàm tái sử dụng]

### Học gì:
- tách `main.py` thành nhiều file
- tạo `services/`, `schemas/`, `prompts/`
- clean code [mã sạch] cơ bản cho AI project
- vì sao AI Engineer phải tổ chức code sớm

### Kết quả cuối buổi:
- project gọn hơn
- dễ mở rộng hơn cho các ngày sau

---

## Day 4
### Chủ đề:
Input validation [kiểm tra đầu vào] + error handling [xử lý lỗi]

### Học gì:
- lỗi thiếu API key
- lỗi empty input [đầu vào rỗng]
- lỗi model output sai schema
- try/except
- debug mindset [tư duy debug]

### Kết quả cuối buổi:
- app bớt mong manh
- hiểu cách làm app ổn định hơn

---

## Day 5
### Chủ đề:
Batch processing [xử lý hàng loạt]

### Học gì:

- đọc từ file `.txt` hoặc `.json`
- lưu nhiều output
- tư duy pipeline [đường ống xử lý]

### Kết quả cuối buổi:
- app có thể chạy trên nhiều input, không chỉ 1 đoạn text

## Day 6

### Học gì:

- so sánh kết quả

### Kết quả cuối buổi:

## Day 7

Mini review [ôn tập nhỏ] + mini project 1

### Kết quả cuối buổi:
