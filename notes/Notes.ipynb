# DAY8 NOTES

Hôm nay tôi học về RAG.

Những gì tôi hiểu:
- RAG là cách kết hợp truy xuất tài liệu liên quan rồi mới đưa cho model tạo câu trả lời
- RAG giúp dùng được dữ liệu riêng và giảm nguy cơ model bịa thông tin
- Pipeline RAG cơ bản gồm load documents, chunking, embeddings, vector store, retrieval và generation

Vì sao phần này quan trọng:
- App hiện tại của tôi mới chỉ xử lý text được đưa trực tiếp vào prompt
- Nếu muốn hỏi đáp trên tài liệu nội bộ hoặc tập tài liệu dài, tôi cần retrieval

Use cases tôi muốn build:
- Hỏi đáp policy nội bộ
- Hỏi đáp data dictionary hoặc schema mô tả
- Trợ lý đọc và tóm tắt nhiều tài liệu nghiệp vụ

# DAY10 NOTES

Hôm nay tôi học về chunking cho RAG.

Những gì tôi hiểu:
- Chunking là chia tài liệu dài thành các đoạn nhỏ hơn để embedding và retrieval hiệu quả hơn
- Chunk size quá nhỏ sẽ làm mất ngữ cảnh
- Chunk size quá lớn sẽ làm retrieval kém chính xác
- Overlap giúp giữ liên kết ngữ cảnh giữa các chunk liên tiếp

Những gì tôi làm được:
- Viết hàm chunk_text_by_words
- Tạo chunks từ file văn bản mẫu
- Lưu chunks ra file JSON
- So sánh nhiều cấu hình chunk size và overlap khác nhau

Điểm tôi học được:
- Chunking là bước rất quan trọng trong RAG
- Chất lượng retrieval phụ thuộc nhiều vào cách chia chunk

# DAY11 NOTES

Hôm nay tôi học về embeddings và similarity search.

Những gì tôi hiểu:
- Embedding là cách biến text thành vector số
- Text có nghĩa gần nhau thường có vector gần nhau
- Retrieval trong RAG thường tìm chunk gần query nhất bằng similarity search
- Cosine similarity là một cách phổ biến để đo độ gần giữa hai vector

Những gì tôi làm được:
- Viết hàm cosine_similarity
- Tạo demo retrieval đơn giản với các vector mẫu
- Sắp xếp chunk theo độ tương đồng với query

Điểm tôi học được:
- Embeddings là cầu nối giữa text và retrieval
- Chunking và embeddings phải đi cùng nhau trong RAG

# DAY12 NOTES

Hôm nay tôi học về vector store và top-k retrieval.

Những gì tôi hiểu:
- Vector store là nơi lưu text chunks, embeddings, và metadata
- Khi có query, hệ thống biến query thành vector rồi tìm các chunk gần nhất
- Top-k retrieval chỉ lấy vài chunk liên quan nhất thay vì toàn bộ tài liệu

Những gì tôi làm được:
- Viết hàm search_top_k
- Tạo một mini vector store bằng Python
- Truy vấn và lấy top-k kết quả gần nhất
- Thử thay đổi query và giá trị k

Điểm tôi học được:
- Vector store là phần trung gian quan trọng giữa chunking và generation
- Metadata rất hữu ích để theo dõi nguồn và lọc dữ liệu
